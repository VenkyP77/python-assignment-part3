```python
# Task 1 — File Read & Write Basics

import os
import sys

# Part A - Write

notes = [
    "Topic 1: Variables store data. Python is dynamically typed.",
    "Topic 2: Lists are ordered and mutable.",
    "Topic 3: Dictionaries store key-value pairs.",
    "Topic 4: Loops automate repetitive tasks.",
    "Topic 5: Exception handling prevents crashes.",
]

# Writing the initial 5 topics from notes
try:
    with open("python_notes.txt", "w", encoding="utf-8") as f:
        for line in notes:
            f.write(line + "\n")
    print("File written successfully.")
except IOError as e:
    print(f"Ugh, something went wrong writing the file: {e}")
    sys.exit(1)

# Appending 2 extra lines
my_extra_notes = [
    "Topic 6: Object Oriented Programming techniques transformed software development.",
    "Topic 7: Agentic AI Workflows have the potential to disrupt established business practices and patterns.",
]

try:
    with open("python_notes.txt", "a", encoding="utf-8") as f:
        for line in my_extra_notes:
            f.write(line + "\n")
    print("Lines appended.")
except IOError as e:
    print(f"Could not append to the file: {e}")
    sys.exit(1)


# Part B - Read

print("\n--- Reading the file ---\n")

all_lines = []

try:
    with open("python_notes.txt", "r", encoding="utf-8") as f:
        all_lines = f.readlines()
except FileNotFoundError:
    print("Something went wrong. The file doesn't exist!")
    sys.exit(1)
except IOError as e:
    print(f"Error reading the file: {e}")
    sys.exit(1)

# print each line with a number in front of it
for i, line in enumerate(all_lines, start=1):
    cleaned = line.rstrip("\n")
    print(f"{i}. {cleaned}")

print(f"\nTotal number of lines: {len(all_lines)}")

# keyword search
print()
keyword = input("Enter a keyword to search for: ").strip()

if not keyword:
    print("You didn't type anything... try running again with an actual word.")
else:
    matches = []
    for line in all_lines:
        if keyword.lower() in line.lower():
            matches.append(line.rstrip("\n"))

    if matches:
        print(f"\nLines containing \"{keyword}\":")
        for m in matches:
            print(f"  -> {m}")
    else:
        print(f"No lines matched \"{keyword}\". Maybe try a different word like 'data' or 'loops'?")
```

```python
# Task 2 — API Integration

import requests
import json

API_BASE = "https://dummyjson.com/products"


def print_table(products, title=""):
    """function for printing products in a tablular format"""
    if title:
        print(f"\n{'========================================================================='}")
        print(f"  {title}")
        print(f"{'========================================================================='}")

    header = f"{'ID':<4}| {'Title':<31}| {'Category':<14}| {'Price':<9}| {'Rating'}"
    separator = f"{'-'*4}|{'-'*32}|{'-'*15}|{'-'*10}|{'-'*7}"
    print(header)
    print(separator)

    for p in products:
        pid = str(p["id"])
        name = p["title"][:30]  # Limiting length of long names so that the table doesn't break
        cat = p["category"][:13]
        price = f"${p['price']:.2f}"
        rating = f"{p['rating']:.2f}"
        print(f"{pid:<4}| {name:<31}| {cat:<14}| {price:<9}| {rating}")


# 1. Fetching 20 products

print("Fetching 20 products from the API...")
resp = requests.get(API_BASE, params={"limit": 20})

data = resp.json()
all_products = data["products"]

print(f"Got {len(all_products)} products back.\n")
print_table(all_products, "All Products (first 20)")


# 2. Filtering ratings >= 4.5 and sort by price descending

good_products = [p for p in all_products if p["rating"] >= 4.5]
good_products.sort(key=lambda x: x["price"], reverse=True)

print_table(good_products, f"Highly Rated Products (rating >= 4.5) — {len(good_products)} found, sorted by price desc")


# 3. Searching for laptops category

print(f"\n{'========================================================================='}")
print("  Laptops Category")
print(f"{'========================================================================='}")

laptop_resp = requests.get(f"{API_BASE}/category/laptops")

if laptop_resp.status_code == 200:
    laptops = laptop_resp.json()["products"]
    for lap in laptops:
        print(f"  - {lap['title']}: ${lap['price']:.2f}")
else:
    print(f"Failed to fetch laptops, status {laptop_resp.status_code}")


# 4. Creating a new product

print(f"\n{'=' * 80}")
print("  Creating a New Product (POST)")
print(f"{'=' * 80}")

new_product = {
    "title": "My Magic Product",
    "price": 420,
    "category": "electronics",
    "description": "A product created via API"
}

post_resp = requests.post(f"{API_BASE}/add", json=new_product)

if post_resp.status_code in (200, 201):
    created = post_resp.json()
    print("Product created successfully! Server response:")
    print(json.dumps(created, indent=2))
else:
    print(f"POST failed with status {post_resp.status_code}")
    print(post_resp.text)
```
